```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM , Bidirectional , GRU
import numpy as np
import random
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
num_sampel = 60
all_arr_train_datasets =[]
all_arr_label_datasets =[]
```

```python
json_files_names = [
    'class_0_normal_datase_1.json', 
    'class_0_normal_datase_2.json',
    'class_0_normal_datase_3.json', 
    'class_0_normal_datase_4.json' ,

'class_1_elbow_up_datase_1.json',
'class_1_elbow_up_datase_2.json',
'class_1_elbow_up_datase_3.json',
'class_1_elbow_up_datase_4.json',

    'class_2_hand_faced_datase_1.json',
    'class_2_hand_faced_datase_2.json',
    'class_2_hand_faced_datase_3.json',
    'class_2_hand_faced_datase_4.json',
]
```

```python
# Repeting times:
for _ in range(4):
    # gener arrays that use to do data agumentation
    agu_arrays = []

    x1_large = np.random.randint(160,370)
    x2_large = np.random.randint(160,370)
    y1_large = np.random.randint(160,370)
    y2_large = np.random.randint(160,370)

    x1_smill = np.random.randint(5,95)
    x2_smill = np.random.randint(5,95)
    y1_smill = np.random.randint(5,95)
    y2_smill = np.random.randint(5,95)

    arr_x_y_large = np.array([x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large,x1_large,y1_large])
    agu_arrays.append(arr_x_y_large)
    arr_x_y_smill = np.array([x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill,x1_smill,y1_smill])
    agu_arrays.append(arr_x_y_smill)
    arr_x_large = np.array([x2_large,0,x2_large,0,x2_large,0,x2_large,0,x2_large,0,x2_large,0,x2_large,0,x2_large,0,x2_large,0,x2_large,0,x2_large,0])
    agu_arrays.append(arr_x_large)
    arr_x_smill = np.array([x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0,x2_smill,0])
    agu_arrays.append(arr_x_smill)
    arr_y_large = np.array([0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large,0,y2_large])
    agu_arrays.append(arr_y_large)
    arr_y_smill = np.array([0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill,0,y2_smill])
    agu_arrays.append(arr_y_smill)


    #===========================================================================
    # Loop to get agumented data  ===> this loop must run two times,
    #  one for adding data,  second for reduce data after re run previouse agu_arrays
    #============================================================================
    for agu_arr in agu_arrays:
        # Loop trough all json files 
        for json_file_name in json_files_names:    
            try:
                print(json_file_name)
                df =  pd.read_json(f'./dataset/{json_file_name}')
                train_column = df.loc[:,'data_rows']
                label_column = df.loc[: , 'training_label']

                # convert all list inside dataset to numpy array:
                arr_dataset = []
                for data_row in train_column:
                    new_data_row = np.array(data_row) + agu_arr # this line must be change every time we re-run after re-generat augented data 
                    arr_dataset.append(data_row)

                # show shpae:
                arr_dataset = np.array(arr_dataset)
                arr_dataset.shape

                # Create temporary tow arrays
                tem_arr_train_dataset = []
                tem_arr_label_dataset = []

                # Basic Setting
                num_package = int(len(arr_dataset)/num_sampel)
                print(num_package)
                start =  0
                end = num_sampel

                # Loop to make Backages
                for _ in range(num_package):
                    curr_arr = np.array(arr_dataset[start:end])
                    if len(curr_arr) == num_sampel:
                        tem_arr_train_dataset.append(curr_arr)
                        start +=  num_sampel
                        end +=  num_sampel

                # Abbend all data to final arrays (all_arr_train_datasets and all_arr_label_datasets)
                for inx , data in  enumerate(tem_arr_train_dataset):
                    all_arr_train_datasets.append(np.array(data))
                    all_arr_label_datasets.append(np.array(label_column[0]))
            except:
                print(f"Not opened : {json_file_name}")

```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(all_arr_train_datasets), np.array(all_arr_label_datasets), test_size=0.10, shuffle=True, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(np.array(x_train), np.array(y_train), test_size=0.10, shuffle=True, random_state=42)
```

```python
model = Sequential()
model.add(GRU(64, input_shape=(x_train.shape[1:]), activation='relu', return_sequences=True))
model.add(Dropout(0.2))

model.add(GRU(64, activation='sigmoid' , return_sequences=True))
model.add(Dropout(0.1))

model.add(GRU(64, activation='sigmoid' , return_sequences=True))
model.add(Dropout(0.1))

model.add(GRU(64, activation='sigmoid'))
model.add(Dropout(0.1))


model.add(Dense(32, activation='sigmoid'))
model.add(Dense(32, activation='sigmoid'))
# model.add(Dense(32, activation='sigmoid'))
model.add(Dropout(0.2))

model.add(Dense(3, activation='softmax'))
```

```python
opt = tf.keras.optimizers.Adam(lr=0.001, decay=1e-6)

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=opt,
    metrics=['accuracy'],
)

history = model.fit(x_train,
          y_train,
          epochs=20,
          validation_data=(x_val, y_val))
```

```python
model.evaluate(x_test ,  y_test)
```
